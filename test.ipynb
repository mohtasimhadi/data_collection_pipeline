```python
import cv2
import matplotlib.pyplot as plt
```

```python
def extract_frames(video_path):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return
    
    frame_count = 0
    frames = []
    while True:
        ret, frame = cap.read()
        
        if not ret:
            break
        frames.append(frame)
        frame_count += 1
    cap.release()
    print(f"Frames extracted: {frame_count}")
    return frames
```

```python
frames1 = extract_frames("1/thread_01_color.h265")
```

```python
frames2 = extract_frames("1/thread_02_color.h265")
```
